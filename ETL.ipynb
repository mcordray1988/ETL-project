```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import json
__count__ = 0
```

```python
url = "http://api.collegefootballdata.com/"
query_url = f"{url}games?year=2019&seasonType=regular&conference=SEC"
```

```python
response = requests.get(query_url).json()
```

```python
print(response)
```

```python
id_num = []
week = []
hometeam = []
awayteam = []
neutral_site = []
homepoints = []
awaypoints = []

```

```python
for x in range(len(response)):
    id_num.append(response[x]["id"])
    week.append(response[x]["week"])
    hometeam.append(response[x]["home_team"])
    homepoints.append(response[x]["home_points"])
    awayteam.append(response[x]["away_team"])
    awaypoints.append(response[x]["away_points"])
    neutral_site.append(response[x]["neutral_site"])
```

```python
game_db = pd.DataFrame({"ID": id_num,
                        "Week": week,
                        "home_team": hometeam,
                        "home_points": homepoints,
                        "away_team": awayteam,
                        "away_points": awaypoints,
                        "neutral_site": neutral_site})
```
